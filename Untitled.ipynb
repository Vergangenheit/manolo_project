```python
import pandas as pd
import numpy as np
from typing import List
import itertools
```

```python
data = './data/TM _Parco_51.xlsx'
```

```python
df = pd.read_excel(data, usecols=list(range(5)))
```

```python
df
```

```python
verbi = df['Verbi'].dropna().unique().tolist()
```

```python
verbi
```

```python
# put column unique data them in lists
lists = {}
for col in df.columns:
    lists[col] = df[col].dropna().unique().tolist()
```

```python
lists
```

```python
list(lists.values())
```

```python
res = list(itertools.product(*list(lists.values()))) 
```

```python
res
```

```python
for r in res:
    assert len(r) == len(df.columns)
```

```python
# transform tuples in lists
res = [list(r) for r in res]
```

```python
# build a df from res
result = pd.DataFrame(res)
```

```python
result
```

```python
result_sentences = [' '.join(r) for r in res]
```

```python
len(result_sentences)
```

```python
result_sentences_set = set(result_sentences)
len(result_sentences_set)
```

```python
assert len([' '.join(r) for r in res]) == len(set(' '.join(r) for r in res))
```
